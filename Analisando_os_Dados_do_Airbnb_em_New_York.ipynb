### **Q4. Há *outliers* presentes?**

Como visto anteriormente, as variáveis `price`, `minimum_nights`, `number_of_reviews`, `reviews_per_month`, `calculated_host_listings_counts` e `number_of_reviews_ltm` apresentaram grande discrepância na visualização dos histogramas, o que poderia indicar a presença de *outliers*. 

Duas técnicas podem ajudar a confirmar essa hipótese:
- por meio de um resumo estatístico da variável;
- por meio da representação gráfica dos dados em *boxplot*.

A primeira técnica foi executada em todos os atributos mencionados acima. Enquanto que, a segunda técnica foi aplicada apenas nas variáveis relevantes ao objetivo desse estudo, ou seja, nas variáveis `price` e `minimum_nights`.

```
# gerar resumos estatísticos
df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 
    'calculated_host_listings_count', 'number_of_reviews_ltm']].describe()
```

Na coluna `price` verifica-se a presença de *outliers* uma vez que a mediana do preço é de 131 dólares e que 75% de todo o conjunto de dados dessa variável está abaixo de 220 dólares, mas o valor máximo é de mais 98 mil dólares. Acrescente-se ainda que, o desvio padrão é bem alto: 919 dólares, o que dá 7 vezes o valor da mediana. Tendo em vista que, por New York ter um dos m² mais caros do mundo, é possível que exista um local que custe quase 100 mil dólares uma diária. Contudo, esse valor prejudica a análise dos dados em conjunto e, por isso, torna-se necessário realizar um tratamento para retirada desses *outliers*.

O mesmo ocorre com o atributo `minimum_nights`. Ou seja, a mediana é 10 noites, enquanto 75% do conjunto de dados exige uma estadia mínima de 30 noites. Contudo, o valor máximo registrado é de 1,250 diárias. O que, nesse caso é um pouco ilógico que alguém exija uma estadia mínima no imóvel de quase 3 anos e meio. Portanto, também é uma variável que necessita de tratamento adequado para que possa ser melhor analisada.

Pode-se observar ainda, essa mesma diferença (entre a mediana, 3º quartil e o valor máximo) também nas demais variáveis analisadas acima.

**Boxplot da variável `price`**

Buscou-se encontrar um preço que pudesse excluir os outliers sem tanto prejuízo ao *dataset*. Para isso, considerou-se que no máximo 1% das entradas totais poderiam ser descartadas.

```
# gerar boxplot
df['price'].plot(kind='box', vert=False, figsize=(15,3))
plt.show()

# visualizar a quantidade de valores acima de 1,700 dólares
print('\n<price>\nValores acima de 1,700 dólares/diária')
print('{} entradas'.format(len(df[df.price > 1700])))
print('{:.2f}%'.format((len(df[df.price > 1700]) / df.shape[0]) * 100))
```

Com o estudo acima, chegou-se ao **valor de corte de 1,700 dólares** para limpar os dados.

**Boxplot da variável `minimum_nights`**

O mesmo procedimento acima foi feito com a variável `minimum_nights`.


```
# gerar boxplot
df['minimum_nights'].plot(kind='box', vert=False, figsize=(15,3))
plt.show()

# visualizar a quantidade mínima de noites que seja abaixo de 90 dias
print('\n<minimum_nights>\nEstadia mínima de até 90 dias')
print('{} entradas'.format(len(df[df.minimum_nights > 90])))
print('{:.2f}%'.format((len(df[df.minimum_nights > 90]) / df.shape[0]) * 100))
```

O valor de corte encontrado é de **90 dias**.

**Tratamento de Dados**

A partir das informações acima removeram-se os *outliers* das variáveis `price` e `minimum_nights`. Para isso, realizou-se uma cópia do conjunto de dados para que nela fossem feitas as alterações necessárias para a melhor análise do conjunto de dados. Ao final, plotaram-se novamente os histogramas.

Removeu-se também a variável `license`, pois não contém valores válidos nela.

Não foram realizados tratamentos nas variáveis `number_of_reviews`, `reviews_per_month` e `number_of_reviews_ltm`, referentes às avaliações pois não julgou-se pertinente para o objetivo desse estudo.

E não foram realizados tratamentos nas variáveis `id`, `host_id`, `latitude`, `longitude` e `calculated_host_listings_counts` uma vez que são variáveis de identificações, geolocalização e quantidade de imóveis de um mesmo anfitrião.

```
# copiar o dataframe
df_clean = df.copy()

# remover os anúncios com preço acima de 1,700 dólares
df_clean.drop(df_clean[df_clean.price > 1700].index, axis=0, inplace=True)

# remover os anúncios com estadia mínima de noites acima de 90 dias
df_clean.drop(df_clean[df_clean.minimum_nights > 90].index, axis=0, inplace=True)

# remover variável 'license'
df_clean.drop('license', axis=1, inplace=True)

# plotar histogramas
df_clean.hist(bins=15, figsize=(15,10));
plt.show()
```

Considerou-se pertinente realizar o levantamento estatístico das variáveis `price` e `minimum_nights` novamente, após a limpeza dos dados, para analisar a distribuição dos dados.

```
# gerar resumos estatísticos
df_clean[['price', 'minimum_nights']].describe()
```

Sobre os preços têm-se uma média, de 185 dólares, e uma mediana, de 130 dólares, ou seja, os valores agora estão próximos. O histograma possui um deslocamento para a esquerda pois, como pode-se ver 75% dos registros possuem o preço de até 216 dólares por estadia. 

Já sobre a quantidade mínima de noites, consegue-se observar agora algo que anteriormente estava oculto: dois picos de valores. Em razão disso, será consultado (sem alterar `df_clean`) um novo histograma, com valor máximo de 40 dias para melhor visualizar e compreender esses movimentos.

```
# gerar novo histograma para 'minimum_nights' com valor máximo de 40 noites
minimum_nights_40 = df_clean.drop(df_clean[df_clean.minimum_nights > 40].index, axis=0)
minimum_nights_40.minimum_nights.hist(bins=15, figsize=(15,10));
plt.show()
```

```
# contagem dos dados da variável 'minimum_nights_40' (até 40 dias)
# mostrar apenas os 10 primeiros valores (com maiores contagens)
minimum_nights_40.minimum_nights.value_counts()[:10]
```

```
# mostrar os valores acima em porcentagens
(minimum_nights_40.minimum_nights.value_counts()[:10] / minimum_nights_40.shape[0]) * 100
```

Claramente agora é possível afirmar que os dados concentram-se entre 1 e 3 noites, bem como entre 30 e 31 dias. Em resumo, **44% dos anúncios exigem uma estadia mínima de 30 diárias**, e logo em seguida estão as estadias de 1 a 3 noites que representam 40% do total das locações.

### **Q6. Qual a correlação existente entre as variáveis?**

Verificou-se aqui se existem variáveis que tenham relação umas com as outras. Essa relação é determinada pelo coeficiente de correlação que, em um número de -1 a 1, indica a força e a direção do relacionamento entre as variáveis.

O procedimento abaixo consiste em criar uma matriz de correlação para gerar, em seguida, um mapa de calor, onde é mais fácil perceber a relação entre as variáveis, caso elas existam.

```
# criar uma matriz de correlação
corr = df_clean[['price', 
                 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 
                 'calculated_host_listings_count', 'availability_365', 
                 'number_of_reviews_ltm']].corr()

# mostrar a matriz de correlação
display(corr)
```

```
# plotar o heatmap das correlações
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True);
```

Quanto mais próximo de 1 ou de -1, mais forte é a correlação entre as variáveis. Enquanto que, mais próximo de 0, indica uma ausência de correlação.
Logo, nota-se que a única correlação mais forte, com um coeficiente igual a 0.83, se dá nas seguintes variáveis `reviews_per_month` e `number_of_reviews_ltm`. O que tem razão uma vez que ambas tratam-se da mesma métrica, ou seja, a quantidade de *reviews* que o imóvel obteve, apenas mudando a métrica: uma mede pelo tempo de 1 mês e a seguinte, pelo período dos últimos 12 meses.

*   apartamento/casa inteira

### **Q8. Qual a localidade mais cara do *dataset*?**



Buscou-se identificar os bairros mais caros e os mais baratos para alugar imóveis na cidade de New York. Desse modo, pode-se ter em mente os bairros mais disputados da cidade, bem como os mais econômicos. 

A análise foi realizada primeiro, com a variável `neighbourhood_group` que setoriza o imóvel entre os 5 bairros de New York e, depois, com a variável `neighbourhood`, apenas naqueles bairros considerados mais caros, para checar a variação de preço nessas regiões. Isso tudo levando em consideração o agrupamento dos bairros de acordo com a média de preços de locação.

```
# ver preços médios por 'neighbourhood_group'
df_clean.groupby(['neighbourhood_group']).price.mean().sort_values(ascending=False)
```

De forma mais ampla, identificam-se como **os bairros mais econômicos o Bronx, seguido pelo Queens**. E, **os bairros mais caros: Manhattan e Brooklyn**. Dessa forma, esses dois últimos serão analisados mais profundamente.

```
# ver preços médios por 'neighbourhood' em Manhattan

# criar variável com dados de Manhattan
manhattan = df_clean.loc[df_clean['neighbourhood_group'] == 'Manhattan']

# agrupar por bairros específicos, de acordo com a média de preços
# mostrar os 10 mais caros
manhattan.groupby(['neighbourhood']).price.mean().sort_values(ascending=False)[:10]
```

```
# mostrar as 10 regiões mais baratas em Manhattan
manhattan.groupby(['neighbourhood']).price.mean().sort_values()[:10]
```

```
# ver preços médios por 'neighbourhood' no Brooklyn

# criar variável com dados do Brooklyn
brooklyn = df_clean.loc[df_clean['neighbourhood_group'] == 'Brooklyn']

# agrupar por bairros específicos, de acordo com a média de preços
# mostrar os 10 mais caros
brooklyn.groupby(['neighbourhood']).price.mean().sort_values(ascending=False)[:10]
```

```
# mostrar as 10 regiões mais baratas no Brooklyn
brooklyn.groupby(['neighbourhood']).price.mean().sort_values()[:10]
```

Na ilha de Manhattan, a região do *Theater District* e de *Tribeca* são as mais caras, com um valor de aluguel por noite de cerca de 505 e 430 dólares, respectivamente. Já as vizinhanças mais econômicas são: *Inwood*, *Marble Hill* e *Washington Heights*, onde os preços variam de 88 a 112 dólares uma diária.
